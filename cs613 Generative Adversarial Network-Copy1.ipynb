```python
import matplotlib.pyplot as plt
from numpy import genfromtxt
import numpy as np
data=genfromtxt('mnist_train.csv',delimiter=',')
test_data=genfromtxt('mnist_test.csv',delimiter=',')
ytrain = data[:,0]
y_test=test_data[:,0]
xtrain = data[:,1:data.shape[1]]
x_test = test_data[:,1:test_data.shape[1]]
x_train=[]
y_train=[]
for i in range(ytrain.shape[0]):
    if ytrain[i]==9:
        x_train.append(xtrain[i])
        y_train.append(ytrain[i])
x_train = np.array(x_train)
y_train = np.array(y_train)        
```

```python
def sigmoid(sig):
    y=1./(1.+ np.exp(-sig))
    return y
```

```python
def dlrelu(x, alpha=1e-2):
    dx = np.ones_like(x)
    dx[x < 0] = alpha
    return dx
```

```python
def gradients_dis(a0_d,firstReal,firstFake,x_real,dz0_real,x_fake,iFake):
    global WmidD
    global disW
    global disB
    global b1_d
    dW1 = np.dot(a0_d.T, firstReal) + np.dot(a0_d.T, firstFake)
    db1 = np.sum(firstReal, axis=0)+ np.sum(firstFake, axis=0)
    dW0 =np.dot(x_real.T, dz0_real)+np.dot(x_fake.T, iFake)
    db0 = np.sum(dz0_real, axis=0) + np.sum(iFake, axis=0)
    disW -= lr * dW0
    disB -= lr * db0
    WmidD -= lr * dW1
    b1_d -= lr * db1
```

```python
def gradients_gen(z,dgrad1,Wgen1,z0_g,aSig,dz0_d,disW,grad1):
    global Wgen0
    global b_geni
    global b_genl
    Wgen0 -= lr * np.dot(z.T, ((np.dot((dgrad1), Wgen1.T)) * dlrelu(z0_g, alpha=0)))
    b_geni -= lr * np.sum(((np.dot(dgrad1, Wgen1.T)) * dlrelu(z0_g, alpha=0)), axis=0)
    Wgen1 -= lr * np.dot(aSig.T, (np.dot(dz0_d, disW.T) * pow((1. - np.tanh(grad1)),2))) 
    b_genl -= lr * np.sum((np.dot(dz0_d, disW.T) *pow((1. - np.tanh(grad1)),2)), axis=0)
```

```python
def fake_data(x_train):
    x_fake=[]
    for a in range(60):
        mean=np.mean(x_train[a])
        std=np.std(x_train[a])
        x_fake.append(np.random.normal(loc=mean, scale=std,size=100))
    x_fake=np.array(x_fake)
    return x_fake
```

```python
def real_dis(mid):
    a0_d=mid
    a0_d[a0_d<0] =0
    z1_d = np.dot(a0_d,WmidD) +b1_d
    a1_d=1.0 / (1.0 + np.exp(-(np.dot(a0_d,WmidD) +b1_d)))
    return z1_d,a1_d
```

```python
def fake_dis(mid):
    a0_d=mid
    a0_d[a0_d<0] =0
    z1_d = np.dot(a0_d,WmidD) +b1_d
    a1_d=1.0 / (1.0 + np.exp(-(np.dot(a0_d,WmidD) +b1_d)))
    return a0_d,z1_d,a1_d
```

```python
from matplotlib.animation import FuncAnimation
def train(x_train,y_train,lr):
        img=[]
        x_train = (x_train.astype(np.float32) - 127.5) / 127.5
        filenames = []
        for epoch in range(50):
            for i in range(50):
                d=np.random.randint(len(x_train), size=60)
                x_real=x_train[d]
                z=fake_data(x_train)
                sigGrad = np.dot(z,Wgen0) +b_geni
                aSig=sigGrad
                aSig[aSig<0] =0
                grad1 = np.dot(aSig,Wgen1) + b_genl
                x_fake = np.tanh(grad1)
                mid = np.dot(x_real,disW) + disB
                z1_d_real,a1_d_real = real_dis(mid)
                z0_d= np.dot(x_fake,disW) + disB
                a0_d,z1_d_fake, a1_d_fake= fake_dis(mid)
                firstReal = (-1. / (a1_d_real + 1e-8))* (sigmoid(z1_d_real)* (1. - sigmoid(z1_d_real))) 
                iReal = np.dot(firstReal,WmidD.T) * dlrelu(z0_d)
                firstFake = ( 1. / (1. - a1_d_fake + 1e-8)) * (sigmoid(z1_d_fake)* (1. - sigmoid(z1_d_fake)))
                iFake = np.dot(firstFake,WmidD.T) * dlrelu(z0_d, alpha=0)
                db0_fake = np.sum(iFake, axis=0)
                gradients_dis(a0_d,firstReal,firstFake,x_real,iReal,x_fake,iFake)
                da1_d = -1.0 / (a1_d_fake + 1e-8)  
                dz1_d = da1_d *  (sigmoid(z1_d_fake)* (1. - sigmoid(z1_d_fake)))
                dz0_d = np.dot(dz1_d, WmidD.T) * dlrelu(z0_d)
                dgrad1 = np.dot(dz0_d, disW.T) * (1. - np.tanh(grad1) ** 2)
                gradients_gen(z,dgrad1,Wgen1,sigGrad,aSig,dz0_d,disW,grad1)        
                maxL=-np.log(a1_d_real) - np.log(1 - a1_d_fake)
                m=np.argmax(maxL)
                m=int(m)
            imag = np.reshape(x_fake, (60, 28,28))
            images = np.reshape(x_fake, (60, 28,28))
            fig = plt.figure(figsize=(4, 4))
            #if epoch % display_epochs == 0:
            plt.imshow(images[m] * 127.5 + 127.5, cmap='gray')
            filenames.append(f"./cs613{epoch}.png")
            plt.savefig(f"./cs613{epoch}.png")
        video_save(filenames)
```

```python
from numpy import random
def weight_initialization():
    Wgen0 = np.zeros((100,128))
    for i in range(Wgen0.shape[0]):
        for j in range(Wgen0.shape[1]):
            Wgen0[i][j]=np.random.randn()* np.sqrt(2. /100)
    b_geni = np.zeros((1,128))  
    Wgen1 = np.zeros((128,784))    
    for i in range(Wgen1.shape[0]):
        for j in range(Wgen1.shape[1]):
            Wgen1[i][j]=np.random.randn()*np.sqrt(2. /128)                
    b_genl = np.zeros((1,28 ** 2))  
    disW = np.zeros((28 ** 2,128))    
    for i in range(disW.shape[0]):
        for j in range(disW.shape[1]):
            disW[i][j]=np.random.randn()*np.sqrt(2. /28 ** 2)               
    disB = np.zeros((1,128))  
    WmidD = np.random.randn(128, 1)                
    for j in range(WmidD.shape[0]):
        WmidD[j][0]=np.random.randn()*np.sqrt(2. /128) 
    b1_d = np.zeros((1, 1))  
    return Wgen0,b_geni,Wgen1,b_genl,disW,disB,WmidD,b1_d
```

```python
import imageio
import matplotlib.pyplot as plt
import numpy as np
display_epochs=5

lr = .001
Wgen0,b_geni,Wgen1,b_genl,disW,disB,WmidD,b1_d=weight_initialization()
train(x_train, y_train,lr)
```
